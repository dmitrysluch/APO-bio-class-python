#### Нафига, а главное зачем нужно программирование
+ Посчитать что нибудь несчитаемое - первые компьютеры использовались чтобы решать уравнения, чтобы рассчитывать траектории снарядов
+ Делать что-то скучное постоянно - есть банк, у банка есть процесс который делают ручками. Вжух и он автоматизирован
+ Делать что-то быстро, люди медленные, а пользователи ждать не будут
+ Обрабатывать много данных, у вас есть табличка на несколько терабайт, и с ней надо что-то сделать

#### Нафига, а главное зачем вам нужно программирование
+ Есть огромное количество удобных инструментов для аналитики - часто написать скрипт быстрее, чем посчитать в том же excel
+ Один раз написал скрипт - можешь как угодно заменять данные, новые результаты появятся мгновенно
+ У программирования выше порог вхождения, но после этого можно реализовать любую функциональность. Если какой-то функциональности нет в Excel то ты проиграл, если чего-то нет в pandas, то это можно написать
