Часто нужно работать с несколькими объектами одновременно. Для этого есть списки и кортежи

```python
# Кортеж
my_tuple = (1, 2, 3, 'собака', 3.1415)
# Список
my_list = [1, 2, 3, 'собака', 3.1415]
print(my_tuple, "Тип:", type(my_tuple))
print(my_list, "Тип:", type(my_list))
```

```python
# Конструкторы list() и tuple()
empty_tuple = ()
other_tuple = tuple()
print(empty_tuple == other_tuple)

empty_list = []
other_list = list()
print(empty_list == other_list)
```

```python
# Получение элементов по индексу
my_tuple = (1, 2, 3, 'собака', 3.1415)
print(my_tuple[0], my_tuple[3])

my_list = [1, 2, 3, 'собака', 3.1415]
print(my_list[0], my_list[3])
```

#### Срезы

```python
a = [1, 2, 3, 4, 5, 6, 7]
# Работает со списками, кортежами, строками
print(a[3:5])
print(a[3:-1])
print(a[::-1])
# Только списки
del a[3:5]
print(a)
a[3:4] = [10, 11]
print(a)
```

```python
# Генераторы списков
print([x * x for x in range(1, 10)])
print([x for x in range(1, 10) if x % 3 == 1])
print(['Fizz' if x % 3 == 1 else x for x in range(1, 10)])
```
