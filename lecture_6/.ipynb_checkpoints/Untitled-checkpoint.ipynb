### Теория

+ Мантисса - значащие цифры числа в двоичной системе 52 бита ~ 16 цифр
+ Экспонента - количество знаков в двоичной записи после запятой 11 бит, значения от −1022 до +1023
+ Значения между $-10^{308}$ и $10^{308}$
+ При сложении и вычитании нужно выравнивать мантиссы
+ Чем меньше число - тем точнее мы можем его представить, относительная погрешность примерно одинаковая

### Относительная и абсолютная погрешность

$\newcommand{eps}{\varepsilon}
\eps_{abs} = |\tilde{a} - a|\\
\eps_{rel} = \frac{|\tilde{a} - a|}{|a|}$
При каждой операции с вещественными числами у нас появляется относительная погрешность. 

$\eps = \max(\eps_a, \eps_b)$  
$\eps << 1$  
$a(1 + \eps_a) + b(1 + \eps_b) = a + b + a\eps_a + b\eps_b \leq (a + b)(1 + \eps)$

$a(1 + \eps_a) \cdot b(1 + \eps_b) = ab(1 + \eps_a + \eps_b + \eps_a\eps_b) \approx ab(1 + 2\eps)$

$\frac{a(1 + \eps_1)}{b(1 + \eps_2)} = \frac{a}{b}(1 + \frac{\eps_1 - \eps_2}{1 + \eps_2}) \leq \frac{a}{b}(1 + 2\eps)$

$a(1 - \eps_a) - b(1 - \eps_b) = a - b - \eps_a a + \eps_b b \leq a - b + 2\max(a, b)\eps$  


И это страшно, потому что $a - b$ может быть во много раз меньше, чем $a$ и $b$

### Что делать
+ Не вычитать близкие числа
+ Использовать decimal и fraction если нужна абсолютная погрешность

```python
from decimal import Decimal
Decimal('1e-10') + 1 - 1 == Decimal('1e-10')
```
