# While

квадраты до 10

```python
i = 1
while i <= 10:
    print(i ** 2)
    i += 1
```

А как переделать чтобы считать корень?
А как переделать, чтобы рисовать пирамидку?
А как переделать чтобы вывести только четные от 0 до 10?

```python
i = 1
while i <= 5:
    print('a')
    i += 1
while i <= 10:
    print('b')
    i += 1
while i <= 15:
    print('c')
    i += 1
```

```python
a = input("Your name: ")
while a != 'end':
    print("Hello", a)
    a = input("Your name: ")
```

```python
while True:
    a = input("Your name: ")
    if a == 'end':
        break
    print("Hello", a)
```

```python
while (a := input("Your name: ")) != 'end':
    print('Hello', a)
```

Посчитать сумму четных

Фича питона - двойное < > <= и >=

```python
i = 1
while i < 10:
    if 40 <= i ** 2 <= 56:
        print(i)
    i += 1
```

Оператор while else
Вводим 10 чисел. Напечатать номер 2-ки. Если 2 не нашлась, то вывести "2 not found"

```python
i = 0
while i < 10:
    num = int(input())
    if num == 2:
        print(i + 1)
        break
    i += 1
else:
    print("2 not found")
```
